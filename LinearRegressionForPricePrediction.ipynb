```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import numpy as np
%matplotlib inline
```

```python
housing_data = pd.read_csv("datasets/housing.csv")

housing_data.sample(5)
```

```python
housing_data.dropna(inplace=True)
```

```python
housing_data.loc[housing_data['median_house_value']==500001].count()
```

```python
housing_data.drop(housing_data.loc[housing_data.median_house_value==500001].index, inplace= True)
```

```python
housing_data.shape
```

##### Note 
All features have numerical calues except the ocean_proximity
These values contains categorical or discrete values not continuous values
Convert categorical values to numeric values using one hot encoding

```python
housing_data = pd.get_dummies(housing_data, columns = ['ocean_proximity'])
housing_data.shape
```

```python
housing_data.sample(5)
```

```python
X = housing_data.drop('median_house_value', axis=1)
y = housing_data['median_house_value']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

#### Note


##### Normalization
Normalization scales all numeric features to be between 0 and 1. Havine features in the same scale can vastly improve the performance of your model

```python
from sklearn.linear_model import LinearRegression 

linear_model = LinearRegression(normalize=True).fit(X_train, y_train)
```

```python
linear_model.score(X_train, y_train)
```

```python
predictors = X_train.columns
```

```python
print(predictors)
```

```python
coef = pd.Series(linear_model.coef_, predictors).sort_values()

print(coef)
```

```python
 y_pred = linear_model.predict(X_test)
```

```python
df_pred_actual = pd.DataFrame({'predicted': y_pred, 'actual' : y_test})

df_pred_actual.head(10)
```

```python
from sklearn.metrics import r2_score

print('Testing score : ',r2_score(y_test, y_pred))
```

```python
df_pred_actual_sample = df_pred_actual.sample(100)
df_pred_actual_sample = df_pred_actual_sample.reset_index()
```

```python
plt.figure(figsize=(20,10))

plt.plot(df_pred_actual_sample['predicted'], label="Predicted")
plt.plot(df_pred_actual_sample['actual'], label='Actual')

plt.legend()
plt.show()
```
